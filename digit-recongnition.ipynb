```python
from ultralytics import YOLO

model = YOLO('yolov8l.pt')

results=model.train(
    data='/kaggle/input/digit0to9/data.yaml',
    epochs=200,
    imgsz=640,
    batch=16,
    name='yolo_custom_model',
    workers=8,
    patience=100
)
```

```python
import os
import random

path = '/kaggle/input/digit0to9/test'
images = os.listdir(path)


image_files = [f for f in images ]

if not image_files:
    print(f"No image files found in {path}")
else:
   
    random_image_filename = random.choice(image_files)
    full_image_path = os.path.join(path, random_image_filename)
```

```python
# THIS IS YOLO 8m Model
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/yolo_custom_model/weights/best.pt")

model.predict(
    source=f"{full_image_path}",
    save=True
)
```
